```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sympy import *
import mpmath
```

```python
values = []
ts = np.arange(0.7, 1.6, 0.1)
for t in ts:
    Smth = lambda x: 2 * 3.1415 * (1 - exp(-t * (x ** (-12) - x ** (-6)))) * x ** 2
    values.append(float(mpmath.quad(Smth, [0, np.inf], method='tanh-sinh')))
```

```python
plt.figure(figsize=(16,16))
plt.plot(ts, values, marker='o')
plt.plot(ts, [0] * len(ts))
plt.show()
```

```python
values = []
ts = np.arange(0.5, 2, 0.1)
for t in ts:
    Smth = lambda x: 2 * 3.1415 * (1 - exp(-t * (x ** (-12) - x ** (-6)))) * x ** 2
    values.append(float(mpmath.quad(Smth, [0, np.inf], method='gauss-legendre')))
```

```python
plt.figure(figsize=(16,16))
plt.plot(ts, values, marker='o')
plt.show()
```
